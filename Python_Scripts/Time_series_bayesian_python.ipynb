```python
import pandas as pd
import numpy as np
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Read in model data - round level (37,676)
model_data = pd.read_csv('C:\\KF_Repo\\PGA_Golf\\Tournament_level_model\\Data_manipulation\\model_data.csv')


# Group by count using pandas groupby()
grouped_data = model_data.groupby(['tournament id', 'Round'])['Round_total'].mean().reset_index()

# Rename columns
grouped_data = grouped_data.rename(columns={"tournament id": "tournament id", 
                                            "Round": "Round",
                                            "Round_total": "Avg_Score"})

# Round Avg Score to 2 decimal places (same as strokes gained)
grouped_data['Avg_Score'] = grouped_data['Avg_Score'].round(2)

# Merge dataframes by 'tournament.id' and 'Round'
add_avg = pd.merge(model_data, grouped_data, on=['tournament id', 'Round'])


# Add difference - put same format as strokes gained
# Negative is bad, positive is good
add_avg['Round_sg'] = add_avg['Avg_Score'] - add_avg['Round_total']

# Filter data for players that you want to analyze
my_players = ['Seamus Power', 'Tony Finau']
mydata2 = add_avg[add_avg['player'].isin(my_players)]

# Convert date to datetime format
mydata2['date'] = pd.to_datetime(mydata2['date'])

# Add in a column for date of round
mydata2['date_round'] = mydata2['date'] + pd.to_timedelta(mydata2['Round'] - 4, unit='D')

# Find the earliest date
earliest_date = mydata2['date_round'].min()

# Calculate the time column
mydata2['time'] = (mydata2['date_round'] - earliest_date).dt.days

# Create a sequence of unique dates and assign corresponding time values
unique_dates = mydata2['date_round'].unique()
date_to_time_mapping = pd.DataFrame({'date_round': unique_dates, 'time_2': np.arange(len(unique_dates))})

# Merge the mapping with the original dataframe
mydata2 = pd.merge(mydata2, date_to_time_mapping, on='date_round', how='left')

# Concatenate columns with "_"
unique_tr = mydata2[['tournament name','date_round', 'Round']].drop_duplicates()
unique_tr['cr'] = unique_tr['tournament name'].astype(str) + "_"+ unique_tr['date_round'].astype(str) + "_" +"R"+ unique_tr['Round'].astype(str)

# Concatenate columns with "_"
unique_tourn = mydata2[['tournament name', 'date']].drop_duplicates()
unique_tourn['tourn'] = unique_tourn['tournament name'].astype(str) + "_" + unique_tourn['date'].astype(str)

# Create additional dataframe before filter
mydata_all = pd.merge(mydata2, unique_tr, on=['tournament name','date_round', 'Round'], how='left')
mydata_all = pd.merge(mydata_all, unique_tourn, on=['tournament name', 'date'], how='left')

# Keep using mydata2
mydata2 = pd.merge(mydata2, unique_tr, on=['tournament name','date_round', 'Round'], how='left')
mydata2 = pd.merge(mydata2, unique_tourn, on=['tournament name', 'date'], how='left')


# Filter train and test data
train_data = mydata_all[mydata_all['date_round'] <= "2020-08-30"]
test_data = mydata_all[mydata_all['date_round'] > "2020-08-30"]

# Perform Regression on training data
reg_data = train_data

# Filter by player
power = reg_data[reg_data['player'] == "Seamus Power"]

# Order by date round
power = power.sort_values(by='date_round')

# Create a time series object
ts_data = pd.Series(power['Round_sg'].values, index=power['date_round'])

## partition into train and test
train_series = ts_data[:80]
test_series = ts_data[80:103]
```

```python
import numpy as np
import statsmodels.api as sm
from scipy.stats import chi2, multivariate_normal

# Fit a linear regression model

y = train_series.values
X = train_exog[:,0]
model = sm.OLS(y, X)
results = model.fit()

# Print results
# r squared is 0 - as its just taking the mean of the data (total sum of squares)
# Residual sum of squares is the same as total so rsq = 0 (1 - rss/tss)
print(results.summary())

# Use classical Regression to estimate the below 3 things
# 1 The estimated parameters beta
# 2 The unscaled covariance matrix (need to sqrt this to get std error)
# 3 The residual variance (This is sum squared errors (ssr) / n-p_

# get number of observations
n = np.shape(y)[0]
# get number of predictors
k = np.shape(results.params)[0]

# 3 estimates
# beta hat is estimate from model
beta_hat = results.params
# unscaled covariance matrix
# When only 1 predictor, this is the variance in beta
# This is the scaled version - already multipled by results.scale
v_beta = results.cov_params()
# Residual variance - so sqrt to get sigma hat
sigma_hat = np.sqrt(results.scale)

# Set 1k sims
n_sims = 1000

# 1 sigma for each sim
sigma = np.zeros(n_sims)
# k betas for each sim (n_sims rows and k columns)
beta = np.zeros((n_sims, k))
# n observations for each sim
y_sim = np.zeros((n_sims, n))

# Create x matrix to multiply by
# set it as k rows and n columns
# Mattirx multiplication will n_sims x K by k x n - result is n_sims x n
# will be multiplying by 1 row of beta - which is length k
# x_tilde * beta[s,:] should be of length n
x_tilde = np.ones((k, n))

# Randomly drawing from chi squared distribution with dof = 79
# This has a mean of 79 with an sd of 12.5
# This seems to provide some variance around using sigma hat (residual variance)
# Then given a random draw of sigma, simulate Beta, with mean beta hat
# and variance matrix 
# Beta Range in OLS was -1.6 to -0.16
# I removed sigma for now - think it may apply with additional variable
for s in range(n_sims):
    # Calculate sigma[s]
    df = n - k
    sigma[s] = sigma_hat * np.sqrt(df / chi2.rvs(df))
   
    # Calculate beta[s] - with multivariate you have vector of means for each beta
    # If beta hat has 2 values then 2 values produced for beta
    # diagonal elements of covariance are variances of beta
    # The distribution of a vector of linear regression coefficients is multivariate normal
    beta[s, :] = multivariate_normal.rvs(mean=beta_hat, cov=v_beta)
    
    # Calculate y - each sim has 80 observations by multiplying by x tilde (n rows k pred)
    # Adding Normal Errors here
    y_sim[s,:] = np.random.normal(np.dot(beta[s, :], x_tilde), scale=sigma[s])
    

# Print means
print('mean beta', np.mean(beta))
print('sd beta', np.std(beta))
print('mean sigma', np.mean(sigma))
print('sd y_sim', np.std(y_sim))

# Compare min values
# This should be set to axis = 1 - so taking min of 80 values
# actual min fits in within 95% range
print('95% min y_sim', np.percentile(np.min(y_sim,axis=1),[2.5,97.5]))
print('min y', np.min(y))


# Compare max values - set axis to 1 - max of 80 values
# max value fits in within 95% range
print('95% max y_sim', np.percentile(np.max(y_sim,axis=1),[2.5,97.5])   )
print('max y', np.max(y))


# Have 1k datasets with 80 observations in each
print('y_sim shape', np.shape(y_sim))

# Get specified ranges for each row
y_sim_95 =  np.percentile(y_sim, [2.5, 97.5], axis=0)
y_sim_68 =  np.percentile(y_sim, [16, 84], axis=0)
y_sim_50 =  np.percentile(y_sim, [25, 75], axis=0)

# Broadcast comparison to find elements between min and max for each row
between_min_max_95 = np.logical_and(y >= y_sim_95[0], y <= y_sim_95[1])
between_min_max_68 = np.logical_and(y >= y_sim_68[0], y <= y_sim_68[1])
between_min_max_50 = np.logical_and(y >= y_sim_50[0], y <= y_sim_50[1])

# Get counts
count_95 = np.sum(between_min_max_95)
count_68 = np.sum(between_min_max_68)
count_50 = np.sum(between_min_max_50)

# get percentages of observations between ranges
print('95 percent ',count_95/np.shape(y)[0])
print('68 percent ',count_68/np.shape(y)[0])
print('50 percent ',count_50/np.shape(y)[0])

# No do same analysis for test

y_test = test_series.values

# Get mean values for ysim - this has 80 observations
mean_95 = np.mean(y_sim_95,axis=1)
mean_68 = np.mean(y_sim_68,axis=1)
mean_50 = np.mean(y_sim_50,axis=1)

# Broadcast comparison to find elements between min and max for each row
# We have a range for each simulation
between_min_max_95_t = np.logical_and(y_test >= mean_95[0], y_test <= mean_95[1])
between_min_max_68_t = np.logical_and(y_test >= mean_68[0], y_test <= mean_68[1])
between_min_max_50_t = np.logical_and(y_test >= mean_50[0], y_test <= mean_50[1])

# Get counts
count_95_t = np.sum(between_min_max_95_t)
count_68_t = np.sum(between_min_max_68_t)
count_50_t = np.sum(between_min_max_50_t)

# get percentages of observations between ranges
print('95 percent test',count_95_t/np.shape(y_test)[0])
print('68 percent test',count_68_t/np.shape(y_test)[0])
print('50 percent test',count_50_t/np.shape(y_test)[0])
```
